PySpark Implementation

```python
# Sample data: list of tuples

data = [('John', 28), ('Anna', 24), ('Mike', 35)]

columns = ['Name', 'Age']

# Creating a DataFrame from the list of tuples

df = spark.createDataFrame(data, schema=columns)

# Displaying the DataFrame

df.display()

# Replaced show with display as it's better for visualisation.
```

Basic DataFrame Operations

```python
# Select the 'Name' column

df.select('Name').display()
```

-- Selecting multiple columns

```python
# Select the 'Name' & 'Age' columns

df.select('Name', 'Age').display()
```

Filtering Data

```python
# Filter rows where Age > 25

df.filter(df['Age'] > 25).display()

# Used .filter() method to filter rows where Age > 25.
```

Sorting Data

```python
# Sort the DataFrame by 'Age'

df.orderBy('Age').display()

# Used .orderBy method to sort the DataFrame by 'Age'.
# Group by 'Age' and count the number of occurrences

df.groupBy('Age').count().display()

# Used .groupBy method to group by 'Age' and .count method to count the number of occurrences.
# Add a new column 'AgeGroup' based on the 'Age' column

from pyspark.sql.functions import when

df = df.withColumn('AgeGroup', when(df['Age'] < 30, 'Young').otherwise('Old'))

# Display the updated DataFrame

df.display()

# Used .withColumn method to add a new column 'AgeGroup' based on the 'Age' column.
# Drop the 'Age' column

df = df.drop('Age')

# After using .orderBy method, I further enhanced the code by respectively using .filter, .groupBy, .withColumn, and .drop methods to perform additional operations.
```

Adding a new column

```python
# Add a new column 'Age After 5 Years'

df = df.withColumn('Age After 5 Years', df['Age'] + 5)

# Show the updated DataFrame

df.display()
```
