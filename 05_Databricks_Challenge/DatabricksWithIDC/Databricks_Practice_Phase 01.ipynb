## 2. Registering DataFrames as a SQL View

## 3. PySpark

## SELECT

```python
# List all tables in the catalog and schema
# tables_df = spark.sql(
#     """
#     SHOW TABLES IN workspace.ecommerce
#     """
# )
# display(tables_df)

# Read the CSV file directly using spark.read.csv
customers_df = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_practice/Customer_Transactions.csv",
    header=True,
    inferSchema=True
)

display(
    customers_df.select(
        "customer_id",
        "country"
    )
)
```

## Filter

```python
display(
    customers_df.filter(customers_df.country == "India")
)
```

## Group By

```python
from pyspark.sql.functions import count as cnt

display(
    customers_df
        .groupBy("country")
        .agg(cnt("*").alias("cnt"))
)
```

## Order By

```python
from pyspark.sql.functions import count

display(
    customers_df
        .groupBy("country")
        .agg(cnt("*").alias("cnt"))
        .orderBy("cnt", ascending=False)
)
```
