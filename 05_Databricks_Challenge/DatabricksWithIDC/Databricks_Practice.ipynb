## Platform setup and basic PySpark commands

```python
# Creating a simple DataFrame:
data = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df = spark.createDataFrame(data, ["product", "price"])
display(df)

# Filter expensive products
display(df.filter(df.price > 1000))
# display(df) provides us with an interactive visualization of our DataFrame. Here, in the tabular format.
```
